# (a) Logistic Regression from Scratch

```python
X_train = np.array([[0.346,0.780], [0.303,0.439],[0.358 ,0.729],[0.602,0.863],[0.790,0.753],[0.611,0.965]])
Y_train = np.array([0,0,0,1,1,1])
```

```python
X_test = np.array([[0.959,0.382], [0.750,0.306],[0.395 ,0.760],[0.823, 0.764],[0.761, 0.874],[0.844 ,0.435]])
Y_test = np.array([0,0,0,1,1,1])
```

```python
def sigmoid(z):
    op = 1 / (1 + np.exp(-z))
    return op
```

```python
def model(w,X):
    z = np.array(w[0] + w[1]*np.array(X[:,0]) + w[2]*np.array(X[:,1]))
    return sigmoid(z)
```

```python
def cost(w, X, Y):
    predict = model(w,X)
    res = -1 * sum(Y*np.log(predict) + (1-Y)*np.log(1-predict))
    return res
```

```python
def gradient(w, X, Y):
    predict = model(w,X)
    g = [0]*3
    g[0] = -1 * sum(Y*(1-predict) - (1-Y)*predict)
    g[1] = -1 * sum(Y*(1-predict)*X[:,0] - (1-Y)*predict*X[:,0])
    g[2] = -1 * sum(Y*(1-predict)*X[:,1] - (1-Y)*predict*X[:,1])
    return g
```

```python
def descent(new_w, prev_w, learning_rate):
    print(prev_w)
    print(cost(prev_w, X, Y))
    j=0
    while True:
        prev_w = new_w
        w0 = prev_w[0] - learning_rate*gradient(prev_w, X, Y)[0]
        w1 = prev_w[1] - learning_rate*gradient(prev_w, X, Y)[1]
        w2 = prev_w[2] - learning_rate*gradient(prev_w, X, Y)[2]
        new_w = [w0, w1, w2]
        print(new_w)
        print(cost(new_w, X, Y))
        if (new_w[0]-prev_w[0])**2 + (new_w[1]-prev_w[1])**2 + (new_w[2]-prev_w[2])**2 <pow(10,-6):
            return new_w
        if j>100: 
            return new_w
        j+=1
```

```python
w=[-1,1.5,0.5]
```

```python
prediction = model(w,X_test)
```

```python
y_predict = prediction
y_predict
```

```python
w = descent(w,w,0.1)
```
